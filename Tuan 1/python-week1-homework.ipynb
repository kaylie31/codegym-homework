```python
# Ve hinh tron

import turtle

#đặt kích thước viền cho hình tròn là 5
turtle.pensize (5)

#đặt màu sắc cho viền hình tròn là màu đỏ
turtle.pencolor ("red")

#tuỳ chỉnh điểm bắt đầu vẽ hình tròn
turtle.Turtle().goto(-40, 120)
#đặt màu nền cho hình tròn là màu đỏ
turtle.fillcolor ("red")
turtle.begin_fill()

#đặt bán kính của hình tròn là 150
turtle.circle (150)
turtle.end_fill()
```

```python
import turtle

t = turtle.Turtle()

turtle.bgcolor("cyan")

t.penup()
t.goto(-200,50)
t.pendown()
t.pensize(5)
t.speed(0)

t.color("orange","yellow")
t.begin_fill()
t.circle(50)
t.end_fill()

for i in range(8):
    t.right(90)
    t.forward(40)
    t.backward(40)
    t.left(90)
    t.circle(50,45)

t.penup()
t.goto(0,0)
t.pendown()
t.color("black", "brown")

t.begin_fill()
for i in range(3):
    t.forward(200)
    t.left(120)

t.end_fill()

t.fillcolor("blue")

t.right(90)
t.begin_fill()
for i in range(3):
    t.forward(200)
    t.left(90)

t.end_fill()

t.penup()
t.goto(80,-120)
t.pendown()
t.color("black", "cornsilk")

t.begin_fill()
for i in range(4):
    t.left(90)
    t.forward(82)
t.end_fill()

t.penup()
t.goto(-150,-200)
t.pendown()
t.color("black", "brown")

t.begin_fill()
t.forward(20)
t.right(90)
t.forward(50)
t.right(90)
t.forward(20)
t.right(90)
t.forward(50)
t.end_fill()

t.penup()
t.goto(-125,-150)
t.pendown()
t.color("black", "brown")

t.right(90)
t.begin_fill()
for i in range(3):
    t.forward(80)
    t.right(120)


```
