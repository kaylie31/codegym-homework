```python
#[Bài tập] Weight Conversion GUI using Tkinter
from tkinter import *

def kg_to_gram():
    kg = ent_weight_kg.get()
    gram = float(kg) * 1000
    lbl_result_gram["text"] = f"{round(gram,2)} gram"

def kg_to_pound():
    kg = ent_weight_kg.get()
    pound = float(kg) * 2.20462
    lbl_result_pound["text"] = f"{round(pound, 2)} pounds"

def kg_to_oz():
    kg = ent_weight_kg.get()
    oz = float(kg) * 35.274
    lbl_result_oz["text"] = f"{round(oz,2)} ounce"


window = Tk()
window.title("Weight Converter")
window.resizable(width=False, height=False)

frm_entry_kg = Frame(master=window)
ent_weight_kg = Entry(master=frm_entry_kg,width=10)
lbl_weight_kg = Label(master=frm_entry_kg, text = "kg")

ent_weight_kg.grid(row=0, column=0, sticky="e")
lbl_weight_kg.grid(row=0,column=1,sticky="w")

btn_convert = Button(
    master=window,
    text="Convert",
    command=lambda: [kg_to_gram(), kg_to_pound(), kg_to_oz()]
)

lbl_result_gram = Label(master=window,text="gram")
lbl_result_pound = Label(master=window,text="pound")
lbl_result_oz = Label(master=window,text="ounce")

frm_entry_kg.grid(row=0, column=0, padx=10)
btn_convert.grid(row=0, column=1, pady=10)
lbl_result_gram.grid(row=1, column=0, padx=10)
lbl_result_pound.grid(row=1, column=1, padx=10)
lbl_result_oz.grid(row=1, column=2, padx=10)

window.mainloop()
```

```python
#[Bài tập] Simple GUI calculator using Tkinter

# Bước 1: Import thư viện Tkinter
from tkinter import *

# Bước 2: Khởi tạo biến expression, nơi lưu biểu thức toàn cục
expression = ""

# Bước 3: Khởi tạo hàm cập nhật biểu thức
def update_expression(num):
    global expression
    expression = expression + str(num)
    equation.set(expression)

# Bước 4: Khởi tạo hàm tính biểu thức cuối cùng
def equalpress():
    try:
        global expression
        total = str(round(eval(expression),2)) # round ket qua 2 decimal places
        equation.set(total)
        expression = ""
    except:
        equation.set("Lỗi")
        expression = ""

# Bước 5: Khởi tạo hàm xoá biểu thức
def clear():
    global expression
    expression = ""
    equation.set("")

# Bước 6: Khởi tạo cửa sổ GUI
app = Tk()

# Bước 7: Cài đặt cửa sổ GUI
app.configure(background="light gray")
app.title("Máy tính đơn giản")
app.geometry("350x200")

# Bước 8: Sử dụng StringVar() để khai báo biến equation với giá trị rỗng
equation = tk.StringVar()
equation.set("")

# Bước 9: Khai báo biến expression_field, nơi thể hiện biểu thức và kết quả bên ngoài màn hình GUI
expression_field = tk.Entry(app, textvariable=equation)
expression_field.grid(columnspan=4, ipadx=70)

# Bước 10: Khởi tạo các nút bấm các số từ 0 đến 9
buttons = [
    '1', '2', '3',
    '4', '5', '6',
    '7', '8', '9',
    '0'
]

row, col = 1, 0
for button in buttons:
    tk.Button(app, text=button, command=lambda b=button: update_expression(b)).grid(row=row, column=col)
    col += 1
    if col > 2:
        col = 0
        row += 1

# Bước 11: Khởi tạo các nút toán tử +, -, *, /
operators = ['+', '-', '*', '/']

row, col = 1, 3
for operator in operators:
    if operator == '+':
        plus = tk.Button(app, text=operator, command=lambda o=operator: update_expression(o))
        plus.grid(row=row, column=col)
    elif operator == '-':
        minus = tk.Button(app, text=operator, command=lambda o=operator: update_expression(o))
        minus.grid(row=row, column=col)
    elif operator == '*':
        multiply = tk.Button(app, text=operator, command=lambda o=operator: update_expression(o))
        multiply.grid(row=row, column=col)
    else:
        divide = tk.Button(app, text=operator, command=lambda o=operator: update_expression(o))
        divide.grid(row=row, column=col)
    row += 1

# Bước 12: Khởi tạo nút tính kết quả toán tử
equal = tk.Button(app, text="=", command=equalpress)
equal.grid(row=4, column=3)

# Bước 13: Khởi tạo nút nhập dấu chấm
decimal = tk.Button(app, text=".", command=lambda: update_expression('.'))
decimal.grid(row=4, column=2)

# Bước 14: Khởi tạo nút xoá
clear_button = tk.Button(app, text="C", command=clear)
clear_button.grid(row=4, column=1)

# Bước 15: Gọi vòng lặp sự kiện chính
app.mainloop()
```
