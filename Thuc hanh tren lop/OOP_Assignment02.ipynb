```python
"""
Viết class NhanVien gồm các thuộc tính:
+ Tên
+ Tuổi
+ Địa chỉ
+ Tiền lương
+ Tổng số giờ làm

Và các phương thức:
- Phương thức tạo
- tinh_thuong(): Tính toán và trả về số tiền thưởng của nhân viên theo công thức sau:
Nếu tổng số giờ làm của nhân viên >=200 thì thưởng = lương * 20%
Nếu tổng số giờ làm của nhân viên <200 và >=100 thì thưởng = lương * 10%
Nếu tổng số giờ làm của nhân viên <100 thì thưởng = 0
- print_info() : In ra tất cả các thông tin cá nhân và tiền thưởng của nhân viên

Tiến hành tạo ra hai đối tượng NhanVien và in ra thông tin tương ứng.
Tạo mảng các nhân viên trong công ty.
- xuất tổng lương trả cho nhân viên.
- tìm nhân viên có tiền lương cao nhất.
"""
class NhanVien:
    def __init__(self, ten, tuoi, dia_chi, luong, so_gio_lam):
        self.ho_ten = ten
        self.tuoi = tuoi
        self.dia_chi = dia_chi
        self.tien_luong = luong
        self.tong_so_gio_lam = so_gio_lam
        
    def tinh_thuong(self):
        thuong: int = 0
        
        if self.tong_so_gio_lam >= 200:
            thuong = self.tien_luong * 0.2
        elif self.tong_so_gio_lam >= 100:
            thuong = self.tien_luong * 0.1
#         else:
#             thuong = 0
        
        return thuong
    
    def __str__(self):
        return f"""{self.ho_ten}, {self.tuoi} tuổi, {self.dia_chi}, lương {self.tien_luong},
làm {self.tong_so_gio_lam}, thưởng {self.tinh_thuong()}"""

    def __gt__(self, other):
        return self.tien_luong > other.tien_luong

    

#Demo
nv1 = NhanVien("Tèo", 23, "280 ADV", 1989, 179)
nv2 = NhanVien("Tý", 39, "123 Trương Định", 1234, 201)
nv3 = NhanVien("Lam", 45, "123 Trương Định", 1989, 233)
print(nv1)
str(nv2)

ds_nhan_vien = [nv1, nv2, nv3]
tong_luong_thuong = 0
for nv in ds_nhan_vien:
    print(nv.tien_luong, nv.tinh_thuong())
    tong_luong_thuong += nv.tien_luong + nv.tinh_thuong()

print('Tổng lương thưởng', tong_luong_thuong)

# tìm nhân viên có tiền lương cao nhất
print("HH:", max(ds_nhan_vien))

luong_max = ds_nhan_vien[0].tien_luong
for index in range(1, len(ds_nhan_vien)):
    if ds_nhan_vien[index].tien_luong > luong_max:
        luong_max = ds_nhan_vien[index].tien_luong

ds_nv_max_luong = list(filter(lambda nv: nv.tien_luong == luong_max, ds_nhan_vien))
for item in ds_nv_max_luong:
    print(item)
```
