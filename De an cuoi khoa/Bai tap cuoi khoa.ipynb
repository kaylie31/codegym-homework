```python
from tkinter import *
from tkinter import ttk, StringVar
from tkinter import messagebox
import json
from openpyxl import Workbook

from MHDangNhap import man_hinh_dang_nhap

root = Tk()
root.title("Quản lý học viên")
root.geometry("1200x600")

# Menu chính (main menu)
menu_bar = Menu(root)

# Background color
root.config(bg = "light gray")

# Nhúng menu vào màn hình chính
root.config(menu=menu_bar)

# Log out function
def log_out():
    result = messagebox.askquestion("Log out", "Are you sure?")
    if result == 'yes':
        root.destroy()
    else:
        messagebox.showinfo("Log out", "Log out canceled")

menu_file = Menu(menu_bar, tearoff=False)
menu_file.add_command(label="Login", command=lambda: man_hinh_dang_nhap(root))
menu_file.add_command(label="Logout", command=log_out)
menu_file.add_separator()
menu_file.add_command(label="Close", command=root.destroy)

menu_bar.add_cascade(label="System", menu=menu_file)

def doc_du_lieu_hoc_vien():
    global data_hoc_vien # dạng global
    try:
        with open("data_hoc_vien.json", "r", encoding="utf-8") as myfile:
            data_hoc_vien = json.load(myfile)
                
    except:
        data_hoc_vien = []
    
def xuat_file_xlxs():
    wb = Workbook()
    
    ws = wb.active
    
    # Gán cột tiêu đề
    ws["A1"] = "Mã HV"
    ws["B1"] = "Họ Tên"
    ws["C1"] = "Điện Thoại"
    ws["D1"] = "Quê Quán"
    ws["E1"] = "Ngày Sinh"
    ws["F1"] = "Giới Tính"
    
    for i, item in enumerate(tv.get_children()):
        values = tv.item(item, 'values')
        for j, value in enumerate(values):
            ws.cell(row=i + 2, column=j + 1, value=value)

    wb.save("hoc_vien.xlsx")

# load data
doc_du_lieu_hoc_vien()
    
mnu_export = Menu(menu_bar, tearoff=False)
mnu_export.add_command(label="Export excel file", command=xuat_file_xlxs)
menu_bar.add_cascade(label="Export", menu=mnu_export)

# Chia làm 3 frame: Danh sách (TreeView), thao tác (Button), chi tiết (nhiều widget bên trong...)
frame_danh_sach = ttk.Frame(root)
frame_tac_vu = ttk.Frame(root)
frame_chi_tiet = ttk.Frame(root)
frame_danh_sach.grid(row=0, column=0,padx=5,pady=5)
frame_tac_vu.grid(row=1, column=0, padx=5, pady=10)
frame_chi_tiet.grid(row=2, column=0, padx=5, pady=10)

''' Vùng danh sách '''
my_columns = ('ma_hv', 'ho_ten', 'so_dt', 'que_quan', 'ngay_sinh', 'gioi_tinh') # tuple
tv = ttk.Treeview(root, columns=my_columns, show='headings')
tv.grid(row=0, column=0)
tv.heading(my_columns[0], text='Mã HV')
tv.heading(my_columns[1], text='Họ Tên')
tv.heading(my_columns[2], text='Điện Thoại')
tv.heading(my_columns[3], text='Quê Quán')
tv.heading(my_columns[4], text='Ngày Sinh')
tv.heading(my_columns[5], text='Giới Tính')

for hocvien in data_hoc_vien:
    ma_hv = hocvien.get("ma_hv")
    ho_ten = hocvien.get("ho_ten")
    so_dt = hocvien.get("so_dt")
    que_quan = hocvien.get("que_quan")
    ngay_sinh = hocvien.get("ngay_sinh")
    gioi_tinh = hocvien.get("gioi_tinh")
    # Insert the data into the Treeview
    tv.insert('', END, values=(ma_hv, ho_ten, so_dt, que_quan, ngay_sinh, gioi_tinh))

''' Vùng chi tiết'''

# Function for Adding students
def them_hoc_vien():

    # Create labels and entry fields for user input
    ttk.Label(frame_chi_tiet, text="Mã HV:").grid(row=0, column=0)
    ttk.Label(frame_chi_tiet, text="Họ Tên:").grid(row=1, column=0)
    ttk.Label(frame_chi_tiet, text="Điện Thoại:").grid(row=2, column=0)
    ttk.Label(frame_chi_tiet, text="Quê Quán:").grid(row=3, column=0)
    ttk.Label(frame_chi_tiet, text="Ngày Sinh:").grid(row=4, column=0)
    ttk.Label(frame_chi_tiet, text="Giới Tính:").grid(row=5, column=0)

    ma_hv_entry = ttk.Entry(frame_chi_tiet)
    ho_ten_entry = ttk.Entry(frame_chi_tiet)
    so_dt_entry = ttk.Entry(frame_chi_tiet)
    que_quan_entry = ttk.Entry(frame_chi_tiet)
    ngay_sinh_entry = ttk.Entry(frame_chi_tiet)
    gioi_tinh_var = StringVar()
    gioi_tinh_combobox = ttk.Combobox(frame_chi_tiet, textvariable= gioi_tinh_var, values= ["nam", "nữ"])

    ma_hv_entry.grid(row=0, column=1)
    ho_ten_entry.grid(row=1, column=1)
    so_dt_entry.grid(row=2, column=1)
    que_quan_entry.grid(row=3, column=1)
    ngay_sinh_entry.grid(row=4, column=1)
    gioi_tinh_combobox.grid(row=5, column=1)

    def luu_hoc_vien():
        # Get the input values
        ma_hv = ma_hv_entry.get()
        ho_ten = ho_ten_entry.get()
        so_dt = so_dt_entry.get()
        que_quan = que_quan_entry.get()
        ngay_sinh = ngay_sinh_entry.get()
        gioi_tinh = gioi_tinh_combobox.get()

        # Add the new entry to the Treeview
        tv.insert('', END, values=(ma_hv, ho_ten, so_dt, que_quan, ngay_sinh, gioi_tinh))

        # Update the data_hoc_vien list
        data_hoc_vien.append({
            "ma_hv": ma_hv,
            "ho_ten": ho_ten,
            "so_dt": so_dt,
            "que_quan": que_quan,
            "ngay_sinh": ngay_sinh,
            "gioi_tinh": gioi_tinh
        })

    ttk.Button(frame_chi_tiet, text="Lưu", command=luu_hoc_vien).grid(row=6, column=0, columnspan=2)

# Function for Deleting student:
def xoa_hoc_vien():
    selected_item = tv.selection()
    if selected_item:
        tv.delete(selected_item)

# Function for updating student data:
def cap_nhat_hoc_vien():
    selected_item = tv.selection()
    if selected_item:
        # Get the values of the selected item
        selected_values = tv.item(selected_item, 'values')

        # Create labels and entry fields for editing
        ttk.Label(frame_chi_tiet, text="Mã HV:").grid(row=0, column=0)
        ttk.Label(frame_chi_tiet, text="Họ Tên:").grid(row=1, column=0)
        ttk.Label(frame_chi_tiet, text="Điện Thoại:").grid(row=2, column=0)
        ttk.Label(frame_chi_tiet, text="Quê Quán:").grid(row=3, column=0)
        ttk.Label(frame_chi_tiet, text="Ngày Sinh:").grid(row=4, column=0)
        ttk.Label(frame_chi_tiet, text="Giới Tính:").grid(row=5, column=0)

        
        ma_hv_entry = ttk.Entry(frame_chi_tiet)
        ho_ten_entry = ttk.Entry(frame_chi_tiet)
        so_dt_entry = ttk.Entry(frame_chi_tiet)
        que_quan_entry = ttk.Entry(frame_chi_tiet)
        ngay_sinh_entry = ttk.Entry(frame_chi_tiet)
        gioi_tinh_var = StringVar()
        gioi_tinh_combobox = ttk.Combobox(frame_chi_tiet, textvariable=gioi_tinh_var, values= ["nam", "nữ"])

        ma_hv_entry.grid(row=0, column=1)
        ho_ten_entry.grid(row=1, column=1)
        so_dt_entry.grid(row=2, column=1)
        que_quan_entry.grid(row=3, column=1)
        ngay_sinh_entry.grid(row=4, column=1)
        gioi_tinh_combobox.grid(row=5, column=1)

        # Pre-fill the entry fields with the selected item's values
        ma_hv_entry.insert(0, selected_values[0])
        ho_ten_entry.insert(0, selected_values[1])
        so_dt_entry.insert(0, selected_values[2])
        que_quan_entry.insert(0, selected_values[3])
        ngay_sinh_entry.insert(0, selected_values[4])
        gioi_tinh_var.set(selected_values[5])

        def luu_cap_nhat_hoc_vien():
            # Get the updated input values
            ma_hv = ma_hv_entry.get()
            ho_ten = ho_ten_entry.get()
            so_dt = so_dt_entry.get()
            que_quan = que_quan_entry.get()
            ngay_sinh = ngay_sinh_entry.get()
            gioi_tinh = gioi_tinh_var.get()

            # Update the selected entry in the Treeview
            tv.item(selected_item, values=(ma_hv, ho_ten, so_dt, que_quan, ngay_sinh, gioi_tinh))


        ttk.Button(frame_chi_tiet, text="Lưu Cập Nhật", command=luu_cap_nhat_hoc_vien).grid(row=6, column=0, columnspan=2)

def cap_nhat_va_luu_du_lieu():
    global data_hoc_vien

    # Clear the existing data in data_hoc_vien
    data_hoc_vien = []

    # Iterate through all the items in the Treeview
    for item in tv.get_children():
        values = tv.item(item, 'values')
        ma_hv, ho_ten, so_dt, que_quan, ngay_sinh, gioi_tinh = values

        # Append the updated entry to data_hoc_vien
        data_hoc_vien.append({
            "ma_hv": ma_hv,
            "ho_ten": ho_ten,
            "so_dt": so_dt,
            "que_quan": que_quan,
            "ngay_sinh": ngay_sinh,
            "gioi_tinh": gioi_tinh
        })

    # Save the updated data to the JSON file
    with open("data_hoc_vien.json", "w", encoding="utf-8") as myfile:
        json.dump(data_hoc_vien, myfile)

'''Vùng tác vụ (gồm các button)'''

# Add the "Thêm" button to the task frame
ttk.Button(frame_tac_vu, text="Thêm", command=them_hoc_vien).grid(row=0, column=0, padx=5)
# Add the "Xóa" button to the task frame
ttk.Button(frame_tac_vu, text="Xóa", command=xoa_hoc_vien).grid(row=0, column=1, padx=5)
# Add the "Cập Nhật" button to the task frame
ttk.Button(frame_tac_vu, text="Sửa", command=cap_nhat_hoc_vien).grid(row=0, column=2, padx=5)
# Add the "Update Data" button to update data_hoc_vien
ttk.Button(frame_tac_vu, text="Lưu dữ liệu", command=cap_nhat_va_luu_du_lieu).grid(row=0, column=3, padx=5)

root.mainloop()
```
