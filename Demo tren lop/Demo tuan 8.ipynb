```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

root = Tk()

root.title("Login Form")
root.geometry("400x300")

Label(root, text = "LOGIN FORM", fg = "red", font = "Arial 16").grid(row = 0, column= 0, columnspan= 2)
#bg = background, fg = foreground

Label(root, text= "User Name:").grid(row = 1, column= 0, pady=5)
Entry(root, text= "User Name:").grid(row = 1, column= 1,pady=5)

Label(root, text= "Email:").grid(row = 2, column= 0, pady=5)
Entry(root, text= "Email:").grid(row = 2, column= 1, pady=5)

Label(root, text= "Password:").grid(row = 3, column= 0, pady=5)
Entry(root, text= "Password:").grid(row = 3, column= 1, pady=5)

def handle_login():
    messagebox.showinfo(title ="Hello", message= "Goodbye")

btnLogin = ttk.Button(root, text= "Login", command=handle_login)
btnLogin.grid(row = 4, column=1, pady=5)
btnLogin.bind('<Return>', handle_login)

root.mainloop()
```

```python
# Demo wiget RadioButton
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import showinfo, showerror, showwarning, askquestion

root = Tk()
root.title("DEMO WIGET CONTROL")
root.geometry("400x300")

ttk.Label(root, text="Chon size ao:").pack(fill=X, padx=5, pady=5)

gia_tri_size_chon = StringVar()

sizes_dict = {
    'Small': 'S',
    'Medium': 'M',
    'Large': 'L',
    'Extra Large': 'XL',
    'Extra Extra Large': 'XXL',
}

for (key, value) in sizes_dict.items():
    #showinfo(message =f"{key} ==> {value}")
    ttk.Radiobutton(root, text=key, value=value, variable=gia_tri_size_chon).pack(fill=X,padx=30)

def xu_ly_chon_size():
    showinfo(title="Result", message= f"Choose: {gia_tri_size_chon.get()}")

ttk.Radiobutton(root, text=" Get Result", command=xu_ly_chon_size).pack()

root.mainloop()
```

```python
# Demo Combobox
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import showinfo, showerror, showwarning, askquestion

root = Tk()
root.title("DEMO WIGET CONTROL")
root.geometry("400x300")

ttk.Label(root, text="Nơi sinh:").grid(row=0, column=0, padx=5, pady=5)

gia_tri_chon = StringVar()
tinh_thanhpho = ttk.Combobox(root, textvariable=gia_tri_chon)
tinh_thanhpho.grid(row=0, column=1)
tinh_thanhpho["values"] = ["Hà Nội", "Sài Gòn", "Cần Thơ", "Đà Nẵng"]

def xu_ly_thay_doi_chon(e):
    showinfo(title="Info", message=f"Choose: {gia_tri_chon.get()}")

tinh_thanhpho.bind('<<ComboboxSelected>>', xu_ly_thay_doi_chon)

root.mainloop()
```

```python
# Demo widget: TreeView
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import showinfo, showerror, showwarning, askquestion

root = Tk()
root.title("DEMO WIDGET CONTROL")
root.geometry("600x300")

my_columns = ('ma_nv', 'ho_ten', 'luong') # tuple
tv = ttk.Treeview(root, columns=my_columns, show='headings')
tv.pack()
tv.heading(my_columns[0], text='Mã NV')
tv.heading(my_columns[1], text='Họ tên NV')
tv.heading(my_columns[2], text='Lương')

ds_nhan_vien = [
    ('NV001', 'Teo Nguyen', 1998),
    ('NV002', 'Ty Le', 2004),
    ('NV003', 'Beckham', 4999)

]

for item in ds_nhan_vien:
    tv.insert('', END, values = item)

def xu_ly_chon(e):
    for item in tv.selection():
        selected_item = tv.item(item)["values"]
        showinfo(message=f"{selected_item[0]} {selected_item[1]} {selected_item[2]}")

tv.bind('<<TreeviewSelect>>', xu_ly_chon)
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import showinfo


root = Tk()
root.geometry("400x300")

# Menu chính (main menu)
menu_bar = Menu(root)
# nhúng menu vào màn hình chính
root.config(menu=menu_bar)

# Menu nhóm File
menu_file = Menu(menu_bar, tearoff=False)
menu_file.add_command(label="New")
menu_file.add_command(label="Open")
menu_file.add_command(label="Save")
menu_file.add_command(label="Close")
menu_file.add_separator()
menu_file.add_command(label="Exit", command=root.destroy)

menu_bar.add_cascade(label="File", menu=menu_file)

# Nhóm menu Help
mnu_help = Menu(menu_bar, tearoff=0)
mnu_help.add_command(label="About Jupyter", command=lambda: showinfo(title='ABOUT', message="MY APP"))
menu_bar.add_cascade(label="Help", menu=mnu_help)

# ================Màn hình con=================
def hien_man_hinh_con():
    mh_con = Toplevel(root)
    mh_con.geometry("300x300")
    mh_con.title("MÀN HÌNH CON")

    ttk.Label(mh_con, text="ĐÂY LÀ MÀN HÌNH CON").pack(fill=X, padx=5, pady=5)
    
    mh_con.mainloop()
# ================End Màn hình con==============

#Thêm menuItem: 
# menu_file.add_command(label="Mở MH Con", command=hien_man_hinh_con)

root.mainloop()
```
