```python
import math
import turtle

class Circle:
    radius = 0
    color = ''
    x = 0
    y = 0

    def __init__(self, radius, color, x, y):
        self.radius = radius
        self.color = color
        self.x = x
        self.y = y

    def getRadius(self):
        return self.radius

    def getArea(self):
        return math.pi * self.radius ** 2

    def getPerimeter(self):
        """
        Phương thức tính chu vi đường tròn
        """
        return 2 * math.pi * self.r

    def draw(self):
        """Phương thức vẽ đường tròn"""
        t = turtle.Turtle()
        t.hideturtle()
        t.color(self.color)
        t.penup()
        t.goto(self.x, self.y)
        t.pendown()
        t.circle(self.radius)
        turtle.done()

    def str(self):
        return f"Circle {self.color} R = {self.radius}, S = {self.getArea()}"

#Demo
ht1 = Circle(7, 'blue', 100, 100)
ht2 = Circle(9, 'black', 200, 200)
print(ht1.radius, ht1.getRadius())
print(ht2)
ht1.draw()
```

```python
from math import sqrt
class Diem:
    # Khai bao thuoc tinh (khong bat buoc):
    x: int = 0
    y: int = 0
    label: str = ''

    def __init__(self,label, x , y):
        self.x = x
        self.y = y
        self.label = label
      
    def __str__(self):
        return f"{self.label} ({self.x}, {self.y})"
    
    def dich_chuyen(self, dx, dy):
        self.x += dx
        self.y += dy

    def khoang_cach_toi_diem(self, other: Diem):
        delta_x = self.x - other.x
        delta_y = self.y - other.y
        return sqrt(delta_x**2 + delta_y**2)

    
#Demo
A = Diem('P', 11, 4)
print(A)
B = Diem ('Q', 11, -9)
print(B)
print(B.khoang_cach_toi_diem(A))
```

```python
from math import gcd
class PhanSo:
    tu_so = 1
    mau_so = 1
    def __init__(self, tu, mau):
        self.tu_so = tu
        self.mau_so = mau

    def __str__(self):
        return f"{self.tu_so}/{self.mau_so}"
    
    #Rut gon phan so
    def rut_gon(self):
        ucln = gcd(self.tu_so, self.mau_so)
        self.tu_so = int(self.tu_so / ucln)
        self.mau_so = int(self.mau_so / ucln)
    
    def __add__(self, other):
        tu_so_moi = self.tu_so * other.mau_so + self.mau_so * other.tu_so
        mau_so_moi = self.mau_so * other.mau_so
        ket_qua = PhanSo(tu_so_moi,mau_so_moi)
        ket_qua.rut_gon()
        return ket_qua

    def __mul__(self, other):
        tu_so_moi = self.tu_so * other.tu_so
        mau_so_moi = self.mau_so * other.mau_so
        ket_qua = PhanSo(tu_so_moi,mau_so_moi)
        ket_qua.rut_gon()
        return ket_qua

ps1 = PhanSo(3, 4)
print(ps1)
ps2 = PhanSo (12,14)
ps2.rut_gon()
print(ps2)
print(ps1 + ps2)
print(ps1*ps2)
```

```python
class SoHoc:
    number1 = 0
    number2 = 1

    def __init__(self, number1, number2):
        self.number1 = number1
        self.number2 = number2

    def print_info(self):
        print(f"Number 1: {self.number1}, Number 2: {self.number2}")

    def __str__(self):
        return f"Number 1: {self.number1}, Number 2: {self.number2}"

    def addition(self):
        return self.number1 + self.number2
    
    def sutract(self):
        return self.number1 - self.number2
    
    def multi(self):
        return self.number1 * self.number2
    
    def division(self):
        if self.number2 == 0:
            print("Can't divide by zero")
            return None
        else:
            return self.number1 / self.number2
    
#Demo
s1 = SoHoc(12,0)
print(s1)
print(s1.division())
```
