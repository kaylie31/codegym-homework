```python
#[Thực hành] Xây dựng lớp hình tròn và vẽ

import turtle
import math

class Circle():
    def __init__(self, r, x, y):
        self.r = r
        self.x = x
        self.y = y

    def draw(self):
        t = turtle.Turtle()
        t.hideturtle()
        t.penup()
        t.goto(self.x, self.y)
        t.pendown()
        t.circle(self.r)
        turtle.done

    def area(self):
        return math.pi * self.r ** 2
    
    def perimeter(self):
        return math.pi * self.r * 2
    
c = Circle(-100,50,0)
c.draw()
print("Area", c.area())
print("Perimeter", c.perimeter())
```

```python
#[Thực hành] Python program to build flashcard using class in Python

import random

class flashcard:
    def __init__(self):

        self.animals = {'Con ong': 'bee',
                        'Con thỏ': 'rabbit',
                        'Con cua': 'crab',
                        'Con mèo': 'cat',
                        'Con ngựa': 'horse',
                        'Con khỉ': 'monkey',
                        'Con lừa': 'donkey'}
        
    def quiz(self):
        while True:

            vietnamese, english = random.choice(list(self.animals.items()))

            print("Tiếng Anh của '{}' là: ".format(vietnamese))
            user_answer = input()

            if user_answer.lower() == english:
                print("Đúng")
            else:
                print("Sai")

            option = int(input("Nhập 0 để tiếp tục : "))
            if option:
                break
    print("Kết thúc")

fc = flashcard()
fc.quiz()
```

```python
# [Bài tập] Continuous Clock with Python Turtle

import tkinter as tk
import math
import time

class Clock:
    def __init__(self, canvas):
        self.canvas = canvas

    def draw_clock(self):
        # Draw clock face
        self.canvas.create_oval(50, 50, 250, 250, width=2)

        # Draw clock hands
        for i in range(12):
            angle = math.radians(90 - i * 30)
            x1 = 150 - 40 * math.cos(angle)
            y1 = 150 - 40 * math.sin(angle)
            x2 = 150 - 30 * math.cos(angle)
            y2 = 150 - 30 * math.sin(angle)
            self.canvas.create_line(x1, y1, x2, y2, width=2)

    def update_time(self):
        # Get real time
        current_time = time.localtime()
        hour = current_time.tm_hour
        minute = current_time.tm_min
        second = current_time.tm_sec

        # Clear canvas
        self.canvas.delete("all")

        # Draw clock
        self.draw_clock()

        # Draw hour hand
        hour_angle = math.radians(90 - (hour % 12) * 30 - minute * 0.5)
        self.canvas.create_line(150, 150, 150 - 40 * math.cos(hour_angle), 150 - 40 * math.sin(hour_angle), width=4, fill='blue')

        # Draw minute hand
        minute_angle = math.radians(90 - minute * 6)
        self.canvas.create_line(150, 150, 150 - 60 * math.cos(minute_angle), 150 - 60 * math.sin(minute_angle), width=3, fill='green')

        # Draw second hand
        second_angle = math.radians(90 - second * 6)
        self.canvas.create_line(150, 150, 150 - 60 * math.cos(second_angle), 150 - 60 * math.sin(second_angle), width=1, fill='red')

        # Update time every second
        self.canvas.after(1000, self.update_time)

# Create window and canvas
root = tk.Tk()
root.title("Clock")

canvas = tk.Canvas(root, width=300, height=300)
canvas.pack()

# Create Clock object and call update_time function
clock = Clock(canvas)
clock.update_time()

# Start event loop
root.mainloop()
```
